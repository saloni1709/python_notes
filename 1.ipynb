```python
# VARIABLE

a = 10
print(a)

b = 20
print(b)

c = 10.1
print(c)

d = True
print(d)

e = False
print(e)

f = "Saloni"
print(f)

g = None
print(g)
```

```python
a = 5
print(type(a))
print(float(a))     # temporarily converts 'a' to a float
                    # But this does NOT change the type of 'a'
print(a)            # 'a' is still the original integer
print(type(a))
a = float(a)       # Now 'a' is permanently converted to float
print(a)
print(type(a))
```
