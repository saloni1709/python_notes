### List → Mutable
## “Mutable” means — you can change the contents of a list after it’s created.

```python
student = ["Saloni", 144, 99, "Jaipur"]
print(student[0])
student[0] = "Seema"
print(student)
```

# List Slicing
##### Slicing means extracting a part of a list using indexes.
### list_name[start : end : step]
##### start → index where slice begins (included) 
##### end → index where slice stops (excluded ❌)
##### step → how many items to skip (default = 1, but if you put a number, that many items will be skipped each time)


```python
marks = [99,98,97,96,95]
print(marks[1:4])
```

```python
marks = [99,98,97,96,95,94,93,92,91,90]
print(marks[1:6:2])
```

```python
marks = [99,98,97,96,95,94,93,92,91,90]
print(marks[:5])
```

```python
marks = [99,98,97,96,95,94,93,92,91,90]
print(marks[5:])
```

```python
marks = [99,98,97,96,95,94,93,92,91,90]
print(marks[::2])
```

```python
marks = [99,98,97,96,95,94,93,92,91,90]
print(marks[-5:-1])
```

## List Methods

```python
list = [10,20,30,40,50]
list.append(100)     # adds one element at the end
print(list)
```

```python
list = [10,100,300,40,50]
list.sort()     # sorts in ascending order
print(list)
```

```python
list = [10,200,30,400,50]
list.sort(reverse=True)       # sorts in descending order
print(list)
```

```python
list = [10,20,30,40,50]
list.reverse()
print(list)
```

```python
list = [10,20,30,40,50]
list.insert(2, 600)            # insert element at index (idx, el)
print(list)
```

```python
list = [10,20,30,40,20]
list.remove(20)     # remove first occurrence of that element
print(list)
```

```python
list = [10,20,30,40,50]
list.pop(3)        # removes element at particular index
print(list)
```

```python
# WAP to check if a list contains a palindrome of elements.(Hint: use copy() method)
list1 = [1,2,3,2,1] 
list2 = list1.copy()
list1.reverse()
if(list1 == list2):
    print("Palindrome list")
else:
    print("Not a palindrome list")
```

```python
# Store the following values in a list & sort them from "A" to "D"
# ["C", "D", "A", "A", "B", "B", "A"]

list = ["C", "D", "A", "A", "B", "B", "A"]
list.sort()
print(list)
```

```python
# You are given a list of subjects for students. Assume one classroom is required for 1 subject. How many classrooms are needed by all students.
#     "python", "java", "C++", "python", "javascript", "java", "python", "java", "C++", "C"

students = ["python", "java", "C++", "python", "javascript", "java", "python", "java", "C++", "C"]
unique_subjects = set(students)
print(unique_subjects)
print(len(unique_subjects))
```

```python
# WAP to print the elements of a list in a single line.(list is the parameter)
def print_list(lst):
    for item in lst:
        print(item, end=' ')
num = [1, 2, 3, 4, 5]
print_list(num)
```
