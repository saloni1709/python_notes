## A built in data type that lets us create immutable sequences of values
Tuples are written in ( ) brackets.

They can store different data types.

They are immutable (cannot be changed).

```python
tuple = (1,2,3,4)
print(type(tuple))
print(tuple[2])
```

```python
tup = ()
print(tup)
print(type(tup))
```

```python
tup = (1,)
print(tup)
print(type(tup))
```

```python
tup = (1,2,1,3,4,1)
print(tup.index(1))     # Return index of first occurrence
```

```python
tup = (1,2,1,3,4,1)
print(tup.count(1))     # Counts total occurrences
```

```python
# WAP to count the number of students with the "A" grade in the following tuple.
# ("C", "D", "A", "A", "B", "B", "A")

grade = ("C", "D", "A", "A", "B", "B", "A")
print(grade.count("A"))
```
