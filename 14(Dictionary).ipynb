## Dictionary Methods

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
print(student.keys())
print(len(student.keys()))
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
print(student.values())
print(len(student.values()))
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
print(student.items())          # Print all key-value pairs as tuples (returns dict_items object)
print(len(student.items()))     # Print how many key-value pairs (items) are present in the dictionary
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
pairs = list(student.items())
print(pairs[0])
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
print(student.get("name"))
print(student["name"])

print(student.get("name1"))           # None
print(student["name1"])               # Error
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
print(student["Subjects"]["math"])
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
student.update({"city" : "jaipur", "rollNo": 144})
print(student)
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}

new_dict = {
    "city" : "jaipur",
    "rollNo" : 144
}
student.update(new_dict)
print(student)
```

```python
student = {
    "name": "Saloni Goyal",
    "Subjects": {
        "physics": 98,
        "chemistry": 97,
        "math": 99
    },
    "grade": "A"
}
student.update({"name": "Seema"})
print(student)
```
