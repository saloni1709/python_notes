### A set in Python is a collection of unique, unordered and mutable(changeable) elements.

👉 Means:

No duplicate values allowed

Elements are not stored in any particular order

You can perform mathematical operations like union, intersection, etc.

```python
collection = {1,2,3,4, "hello", "world"}
print(collection)
print(type(collection))
print(len(collection))
```

```python
collection = {1,2,2,2, "hello", "world", "hello"}
print(collection)
print(type(collection))
print(len(collection))
```

## Set Methods

```python
collection = {1,2,2,4, "hello", "world", "hello"}
collection.add(100)
print(collection)
```

```python
collection = {1,2,2,4, "hello", "world", "hello"}
collection.add(("Heyy", 200, 300))
print(collection)
```

```python
set = {1,2,3,4,5,6,7,"Hello"}
set.remove(3)
print(set)
```

```python
set = {1,2,3,4,5,6,7,"Hello"}
set.clear()           # Empties the set
print(set)
```

```python
set = {1,2,3,4,5,6,7,"Hello"}
set.pop()         # Removes a rendom value
print(set)
```

```python
set = {1,2,3,4,5,6,7,"Hello"}
print(set.pop())
print(set.pop())
```

```python
set1 = {1,2,3}
set2 = {2,3,4}
print(set1.union(set2))
```

```python
set1 = {1,2,3}
set2 = {2,3,4}
print(set1.intersection(set2))
```
