## Break

break is used inside loops (for or while) to stop the loop immediately, even if the loop condition is still true.

<!-- break is used inside loops (for or while) to stop the loop immediately, even if the loop condition is still true. -->

```python
i = 1
while i<=5:
    print(i)
    if(i==3):
        break
    i+=1
print("end of loop")
```

```python
nums = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)
x = 36
i = 0
while i<len(nums):
    if(nums[i] == x):
        print("Found at index", i)
        break
    else:
        print("Finding...")
        i+=1
print("end of loop")
```

```python
str = "apnacollege"
for char in str:
    if(char == 'o'):
        print("o found")
        break
    print(char)
    print("End")
```

## Continue
continue is used inside loops to skip the current iteration and move to the next one, without stopping the loop completely.

```python
i = 1
while i<=5:
    print(i)
    if(i==3):
        i+=1
        continue
    i+=1
print(i)
```
