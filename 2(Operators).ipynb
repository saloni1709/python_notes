## 1. Arithmetic Operators
Used for basic mathematical operations.

```python
num1 = 10
num2 = 5
print(num1+num2)
print(num1-num2)
print(num1*num2)
print(num1/num2)           # Division → 10 / 5 = 2.0 (always returns float)
print(num1//num2)          # Floor Division → 10 // 5 = 2 (removes decimal part)
print(num1%num2)
print(num1**num2)          # Exponentiation → 10 ** 5 = 100000 (10 raised to power 5)
```

## 2. Comparison (Relational) Operators
Used to compare two values (results are always True or False).

```python
a = 20
b = 10
print(a == b)
print(a != b)
print(a < b)
print(a <= b)
print(a > b)
print(a >= b)
```

## 3. Logical Operators
Used to combine conditional statements.

```python
print((10>9) & (20<10))
print((10>9) and (20<10))
print((10>9) | (20<10))
print((10>9) or (20<10))
print(not(10>9))
print(not(9>10))
```

## 5. Bitwise Operators
Work at the binary (bit) level.

```python
a = 2   # binary: 10
b = 1   # binary: 01

print(a & b)   # AND → 10 & 01 = 00 → 0
print(a | b)   # OR  → 10 | 01 = 11 → 3
print(a ^ b)   # XOR → 10 ^ 01 = 11 → 3
print(~a)      # NOT → -(2+1) = -3
print(~b)      # NOT → -(1+1) = -2
print(a >> b)  # Right shift → 10 >> 1 → 1 (binary 1)
print(a << b)  # Left shift → 10 << 1 → 100 → 4
```

## 6. Membership Operators
Used to test if a value is part of a sequence (like string, list, etc.).

```python
a = "My name is Saloni Goyal."
print("Saloni" in a)     # Checks if "Saloni" exists in string 'a'
                         # Returns True if value exists
print("saloni" in a)     # Checks if "saloni" exists in string 'a'
                         # Returns False if value not exists
```

```python
print("a" in "Saloni")
print("h" in "Saloni")
print("s" in "Saloni")
```
