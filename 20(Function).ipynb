A function is a block of code that runs only when it’s called.
It helps you reuse code, make your program more organized, and reduce repetition.

```python
def sum(a, b):
    s = a+b
    return s
print(sum(3, 4))
print(sum(4,5))
```

```python
def diff(a, b):
    d = a-b
    return d
print(diff(4, 2))
print(diff(9, 8))
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact*=i
    return fact
print(factorial(4))
print(factorial(5))
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact*=i
    print(fact)
factorial(3)
factorial(5)
```

```python
def avg(a, b, c):
    avg = (a+b+c)/3
    return avg
print(avg(1, 2, 3))
```

```python
def prod(a, b=2):
    return a*b
print(prod(3))
print(prod(5))
```

```python
def converter(usd_val):
    inr_val = usd_val*83
    print(usd_val, "USD_val", inr_val, "INR_val")
converter(2)
```
