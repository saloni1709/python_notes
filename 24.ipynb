```python
f = open("24.txt", "a+")
f.write("abc")
print(f.read())
f.close()
```

WITH SYNTAX

```python
with open("24.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
with open("24.txt", "w") as f:

     # Write the string "new data" to the file.
    # The write() method returns the number of characters written.
    data = f.write("new data")
```
