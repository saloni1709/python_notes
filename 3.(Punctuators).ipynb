## Python Punctuators (Delimiters / Special Characters)
##### Punctuators are special characters that define structure or syntax in Python.
##### They include     =   { }   [ ]   ( )   ,   :   .   ;   @   ->   ' '   """ """   #

```python
# Python Punctuators / Delimiters Examples

# =   → Assignment
a = 10           # Assigns 10 to variable 'a'
b = 20
print(a + b)     # 30

# {}  → Curly braces
d = {"name": "Saloni", "age": 20}  # Dictionary
print(d["name"])                     # Saloni

# []  → Square brackets
lst = [1, 2, 3, 4]   # List
print(lst[0])         # 1 → Access first element

# ()  → Parentheses
x = (a + b) * 2      # Grouping expression
print(x)             # 60
print(len(lst))      # Using parentheses for functions

# ,   → Comma
a, b = 5, 6           # Multiple assignment
print(a, b)           # 5 6
lst2 = [1, 2, 3, 4]   # Comma separates list elements
print(lst2)

# :   → Colon
for i in range(3):    # Starts an indented block
    print(i)          # 0 1 2

# .   → Dot
s = "hello"
print(s.upper())      # Dot accesses methods of object → HELLO

# ;   → Semicolon
x = 1; y = 2; print(x + y)  # Multiple statements in one line → 3

# @   → At symbol
# Used for decorators
def decorator(func):
    return func

@decorator
def my_func():
    print("Hello")
my_func()             # Hello

# ->  → Used in function annotations (return type)
def add(a: int, b: int) -> int:
    return a + b
print(add(2,3))       # 5

# ' ' or " " → Single or double quotes
name = 'Saloni'
print(name)           # Saloni

# """ """ → Triple quotes for multi-line string
text = """This is
a multi-line
string."""
print(text)

# # → Comment symbol
# Everything after # is ignored by Python
```
