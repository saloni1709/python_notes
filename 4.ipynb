## Expression Execution

#### String and Numeric Values can operate together with *

```python
a, b = 2, 3
txt = "@"
print(a*txt*b)       # This is an expression being executed:
                     # 1. a * txt → 2 * "@" → "@@"   (string repeated)
                     # 2. "@@" * b → "@@" * 3 → "@@@@@@" (string repeated again)
                     # The expression evaluates to a value → "@@@@@@"
```

#### String & String can operate with +

```python
a, b = "2", 3
txt = "@"
print((a+txt)*b)      # Step 1: a + txt → "2" + "@" → "2@"  (string concatenation)
                      # Step 2: "2@" * b → "2@" * 3 → "2@2@2@"  (string repeated 3 times)
```

#### Numeric Values can operate with all arithmetic operators

```python
a = 10
b = 5
c = 2
print(a+b*c)
```

#### Arithmetic expression with Integer and float will result in float

```python
a,b = 10, 5.0
c = a*b
print(c)
```

#### Result of division operator with two integers will be float

```python
a = 12
b = 6
print(a/b)
```

#### Integer division with float and int will give int displayed as float

```python
a,b = 1.5, 3
c = a/b
print(c)
```

#### Floor gives closest integer, which is lesser than or equal to the float value 
#### Result of(a//b) is same as floor(a/b)

```python
a = 13
b = 5
c = a//b
print(c)
```

```python
a = -13
b = 5
c = a//b
print(c)
```

```python
a = 13
b = -5
c = a//b
print(c)
```

#### Remainder is negative when denominator is negative

```python
a,b = 5,2
c=a%b
print(c)
```

```python
a,b = -5,2
c=a%b
print(c)
```

```python
a,b = 5,-2        # a = 5 (positive numerator), b = -2 (negative denominator)
c=a%b
print(c)
```
