##### A string is a sequence of characters enclosed in quotes.  It can include letters, numbers, and symbols.

### String Concatenation
##### String concatenation means joining two or more strings together.

```python
str1 = "This is a string."
str2 = "We are creating it in python."
print(str1 + str2)
```

## String Repetition
##### You can repeat strings using *:

```python
print("Hi " * 3)
```

## Indexing
##### Each character in a string has an index (starting from 0)

```python
str = "apna college"
print(str[3])
print(str[5])
```

## Slicing
##### You can take a part of a string using slicing [start:end]

```python
str = "Saloni Goyal"
print(str[1:9])   # str[1:9] → means start from index 1 and go till index 8 (9 is not included)
print(str[:4])    # [0:4]
print(str[4:])    # [4:len(str)]
```

```python
str = "Saloni Goyal"
print(str[-3:-1])
```

## String Functions

```python
str = "I am studying python from Apna College"
print(str.endswith("Apna"))
print(str.endswith("ge"))
```

```python
str = "I am studying python from Apna College"
print(str.startswith("I am"))
print(str.startswith("am"))
```

```python
str = "i am studying python from Apna College"
print(str.capitalize())      # Capitalize 1st char
```

```python
str = "I am studying python from Apna college"
str = str.capitalize()
print(str)
```

```python
str = "i am studying python from Apna College"
print(str.replace("from", "to"))               # replace old to new (old, new)
```

```python
str = "I am studying python from Apna College"
print(str.find("t"))          # return index of 1st occurrence
```

```python
str = "I am studying python from Apna College"
print(str.count("am"))
print(str.count("a"))
```

### String → Immutable
## “Immutable” means — you cannot change the value once it’s created.

```python
name = "Saloni"
name[0] = "M" 

# You can’t modify a string’s individual characters.
# If you want to “change” it, you must create a new string.
```

```python
name = "Saloni"
new_name = "M" + name[1:]
print(new_name)   # Maloni
```
